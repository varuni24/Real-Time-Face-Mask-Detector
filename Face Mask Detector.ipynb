## 9) Real Time Implementation

```python
cap= cv2.VideoCapture(0)  
while cap.isOpened():
    ret, frame= cap.read()    
    results=model(frame)
    cv2.imshow('MASK DETECTOR',np.squeeze(results.render())) 
    if cv2.waitKey(10) & 0xFF == ord('q'): 
        break
cap.release()
cv2.destroyAllWindows() 
```
